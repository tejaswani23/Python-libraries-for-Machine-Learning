```python
x=['A','B','C']
y=[2,4,6]
plt.figure(figsize=(4,5),dpi=100)
plt.bar(x,y)
```

```python
x=['A','B','C']
y=[2,4,6]
plt.figure(figsize=(3,4),dpi=100)
bars=plt.bar(x,y)
bars[0].set_hatch('/')
bars[1].set_hatch('o')
bars[2].set_hatch('*')
plt.show()
```

```python
x=['A','B','C']
y=[2,4,6]
plt.figure(figsize=(3,4),dpi=100)
bars=plt.bar(x,y)
patterns=['*','*','*']
for bar in bars:
    bar.set_hatch(patterns.pop(0))
plt.show()
```
